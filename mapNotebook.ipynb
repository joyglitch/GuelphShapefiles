```python
import pandas as pd
import geopandas as gpd
%matplotlib inline
```

```python
buildings = gpd.read_file("shape/buildings.shp")
landuse = gpd.read_file("shape/landuse.shp")
natural = gpd.read_file("shape/natural.shp")
places = gpd.read_file("shape/places.shp")
points = gpd.read_file("shape/points.shp")
railways = gpd.read_file("shape/railways.shp")
roads = gpd.read_file("shape/roads.shp")
waterways = gpd.read_file("shape/waterways.shp")

guelph = gpd.read_file("shape/CityBoundary.shp")
```

```python
busStation = points[points['type'] == 'bus_station']
busStop = points[points['type'] == 'bus_stop']
```

```python
sizeX = 20
sizeY = 20

guelphMap = guelph.plot(figsize=(sizeX,sizeY), color='#F4CE52', alpha= 0.8)

#Make roads and railways
fig1 = roads.plot(figsize=(sizeX,sizeY), ax=guelphMap, linewidth=0.5, color='#757575', label='Roads')
fig2 = railways.plot(figsize=(sizeX,sizeY), ax=fig1, color='#C2C5BB', linestyle='--', label='Railways')

#Make nature and waterways
fig3 = natural.plot(figsize=(sizeX,sizeY), ax=fig2, color='#C6E8C6', label='Natural', alpha=0.75)
fig4 = waterways.plot(figsize=(sizeX,sizeY), ax=fig3, linewidth=1, color='#AADAFF', label='Waterways')

#Make place and point markers
#fig = places.plot(figsize=(sizeX,sizeY), ax=fig, markersize=5, color='#ff9966')
#fig = points.plot(figsize=(sizeX,sizeY), ax=fig3, markersize=5, color='#ff3300')
figBusStation = points.plot(figsize=(sizeX,sizeY), ax=fig4, markersize=5, color='#020887', label='Bus Stops')
figBusStop = points.plot(figsize=(sizeX,sizeY), ax=figBusStation, markersize=5, color='#020887')

#Make buildings and landuses
#fig = landuse.plot(figsize=(sizeX,sizeY), ax=fig, color='#D0FCB3')
fig = buildings.plot(figsize=(sizeX,sizeY), ax=figBusStop, linewidth=1, color='#605770', label='Buildings')
figCopy = buildings.plot(figsize=(sizeX,sizeY), ax=fig3, linewidth=1, color='#605770', label='Buildings')

fig.set_facecolor('#F8F9FA')
fig.plot()
fig.legend()
```
